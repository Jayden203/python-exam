## å®æ“ä»»åŠ¡ 1ï¼šå›å½’æ¨¡å‹ï¼ˆMSELoss + Adamï¼‰

```
# ========================== å®æ“ä»»åŠ¡ 1ï¼šå›å½’æ¨¡å‹ ==========================

import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt

torch.manual_seed(1337)

# -------------------------- Step 1ï¼šæ•°æ®ç”Ÿæˆ --------------------------
X = torch.randn(120, 8)
w = torch.randn(8, 1)
b = torch.randn(1)
noise = 0.5 * torch.randn(120, 1)
y = X @ w + b + noise

print("X shape:", X.shape)
print("y shape:", y.shape)

# -------------------------- Step 2ï¼šæ¨¡å‹å®šä¹‰ --------------------------
class RevenueNet(nn.Module):
    def __init__(self):
        super(RevenueNet, self).__init__()
        self.fc1 = nn.Linear(8, 16)
        self.fc2 = nn.Linear(16, 8)
        self.fc3 = nn.Linear(8, 1)
        self.relu = nn.ReLU()

    def forward(self, x):
        x = self.relu(self.fc1(x))
        x = self.relu(self.fc2(x))
        x = self.fc3(x)
        return x

model = RevenueNet()
print(model)

# -------------------------- Step 3ï¼šæ¨¡å‹é…ç½® --------------------------
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.01)

# -------------------------- Step 4ï¼šè®­ç»ƒå¾ªç¯ --------------------------
epochs = 30
losses = []

for epoch in range(epochs):
    preds = model(X)
    loss = criterion(preds, y)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    losses.append(loss.item())
    print(f"Epoch {epoch+1:02d} | Loss: {loss.item():.6f}")

# -------------------------- Step 5ï¼šæŸå¤±å¯è§†åŒ– --------------------------
plt.plot(losses)
plt.title("Training Loss over Epochs")
plt.xlabel("Epoch")
plt.ylabel("MSE Loss")
plt.show()
```

## å®æ“ä»»åŠ¡ 2ï¼šäºŒåˆ†ç±»æ¨¡å‹ï¼ˆBCELoss + SGDï¼‰

```
# ========================== å®æ“ä»»åŠ¡ 2ï¼šäºŒåˆ†ç±»æ¨¡å‹ ==========================

import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt

torch.manual_seed(1337)

# -------------------------- Step 1ï¼šæ•°æ®ç”Ÿæˆ --------------------------
X = torch.randn(200, 4)
w = torch.tensor([[0.8], [0.6], [0.7], [0.5]])
b = torch.tensor([-1.0])
noise = 0.3 * torch.randn(200, 1)
proba = torch.sigmoid(X @ w + b + noise)
y = (proba > 0.5).float()

print("X shape:", X.shape)
print("y shape:", y.shape)

# -------------------------- Step 2ï¼šæ¨¡å‹å®šä¹‰ --------------------------
class RiskNN(nn.Module):
    def __init__(self):
        super(RiskNN, self).__init__()
        self.fc1 = nn.Linear(4, 8)
        self.fc2 = nn.Linear(8, 1)
        self.relu = nn.ReLU()
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = self.relu(self.fc1(x))
        x = self.sigmoid(self.fc2(x))
        return x

model = RiskNN()
print(model)

# -------------------------- Step 3ï¼šæ¨¡å‹é…ç½® --------------------------
criterion = nn.BCELoss()
optimizer = optim.SGD(model.parameters(), lr=0.05)

# -------------------------- Step 4ï¼šè®­ç»ƒå¾ªç¯ --------------------------
epochs = 15
losses = []

for epoch in range(epochs):
    preds = model(X)
    loss = criterion(preds, y)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    losses.append(loss.item())
    print(f"Epoch {epoch+1:02d}: loss={loss.item():.4f}")

# -------------------------- Step 5ï¼šæŸå¤±å¯è§†åŒ– --------------------------
plt.plot(losses)
plt.title("Classification Training Loss over Epochs")
plt.xlabel("Epoch")
plt.ylabel("BCELoss")
plt.show()
```
